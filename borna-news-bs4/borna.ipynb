```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = 'https://www.borna.news/'
all_news = 'https://www.borna.news/newsstudios/archive/?curp=2&order=order_time&dateRange%5Bstart%5D=-2592000&offset%5Bseen%5D=1135195&offset%5Btime%5D=2021-02-10+07%3A29%3A20&dir=1'
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
a = soup.text
```

```python
links = soup.find_all('a')
list = []
for link in links:
    if link.get('title') != None:
        list.append(link.text.strip().replace('\u200c',''))
        
while '' in list:
    list.remove('')
```
